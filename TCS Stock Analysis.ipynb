```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
#IMPORT DATASET

tm = pd.read_csv('TATAMOTORS.csv')
ts = pd.read_csv('TATASTEEL.csv')
tcs = pd.read_csv('TCS.csv')
```

```python
#CHECKING FOR NULL VALUES

tm.isna().sum()
ts.isna().sum()
tcs.isna().sum()
```

```python
#CHECKING FOR DUPLICATE VALUES

tm.duplicated().sum()
ts.duplicated().sum()
tcs.duplicated().sum()
```

```python

#Working on Data
```

```python
#Converting the “Date” column dtype from object to date

tm.drop(['Trades','Deliverable Volume','%Deliverble'],axis=1,inplace=True)
ts.drop(['Trades','Deliverable Volume','%Deliverble'],axis=1,inplace=True)
tcs.drop(['Trades','Deliverable Volume','%Deliverble'],axis=1,inplace=True)
```

```python
#Dropping unwanted columns

tm['Date']=pd.to_datetime(tm['Date'])
ts['Date']=pd.to_datetime(ts['Date'])
tcs['Date']=pd.to_datetime(tcs['Date'])
```

```python
#Adding 3 more new columns to each of the Dataset

tm['Day']=tm['Date'].dt.day
tm['Month']=tm['Date'].dt.month
tm['Year']=tm['Date'].dt.year

ts['Day']=ts['Date'].dt.day
ts['Month']=ts['Date'].dt.month
ts['Year']=ts['Date'].dt.year

tcs['Day']=tcs['Date'].dt.day
tcs['Month']=tcs['Date'].dt.month
tcs['Year']=tcs['Date'].dt.year
```

```python

#EDA
```

```python
#Price Comparision

plt.figure(figsize=(20,7))

plt.plot(tm['Date'],tm['Open'],color='blue',label='Tata Motors')
plt.plot(ts['Date'],ts['Open'],color='green',label='Tata Steel')
plt.plot(tcs['Date'],tcs['Open'],color='orange',label='TCS')

plt.xlabel('Year')
plt.ylabel('Price')
plt.title('RELATION BETWEEN TATA MOTORS, TATA STEEL AND TCS SHARE PRICE')
plt.legend(title='')
```

```python
#Volume Comparision

fig=plt.figure(figsize=(20,7))

#axes1 for outer graph

axes1=fig.add_axes([.1,.1,.8,.8])
axes1.plot(tm['Date'],tm['Volume'],color='blue',label='Tata Motors')
axes1.plot(ts['Date'],ts['Volume'],color='green',label='Tata Steel')
axes1.plot(tcs['Date'],tcs['Volume'],color='orange',label='TCS')
axes1.set_xlabel('Year')
axes1.set_ylabel('Price')
axes1.set_title('RELATION BETWEEN TATA MOTORS, TATA STEEL AND TCS SHARE VOLUME')
axes1.legend(title='')

#axes2 for inner graph

axes2=fig.add_axes([.2,.5,.3,.3])
axes2.plot(tm['Date'].tail(100),tm['Volume'].tail(100),color='blue',label='Tata Motors')
axes2.plot(ts['Date'].tail(100),ts['Volume'].tail(100),color='green',label='Tata Steel')
axes2.plot(tcs['Date'].tail(100),tcs['Volume'].tail(100),color='orange',label='TCS')
axes2.set_xlabel('Year')
axes2.set_ylabel('Price')
axes2.set_title('PEAK TIME')
```

```python

#Return on Investment(ROI)
```

```python
#Tata Motors ROI

sumTM=0

s1=0

#calcuating total amount invested and number of shares owned of Tata Motors

for i in range(len(tm)):

    if tm.loc[i,'Day']==30:

        sumTM+=tm.loc[i,'Open']

        s1+=1

print("Total Invested in Tata Motors = Rs.",round(sumTM,2))

print("Shares Owned of Tata Motors =",s1)

print("Average Investmentment of 1 share = Rs.",round((sumTM/s1),2))



tm_end=298.2       #last open price of Tata Motors on 2021-04-30

#calculating investment results

result1=round((tm_end*s1)-sumTM,2)

roiTM=round((result1/sumTM)*100,2)


print("Investment Result:")


if result1<0:

    print("Net Unrealised Loss = Rs.",abs(result1))

else:

    print("Net Unrealised Profit = Rs.",result1)



print("Tata Motors ROI from 2000-1-3 to 2021-04-30 =",roiTM,"%")
```

```python
#Tata Steel ROI

sumTS=0

s2=0

#calcuating total amount invested and number of shares owned of Tata Steel

for i in range(len(ts)):

    if ts.loc[i,'Day']==30:

        sumTS+=ts.loc[i,'Open']

        s2+=1

print("Total Invested in Tata Steel = Rs.",round(sumTS,2))

print("Shares Owned of Tata Steel =",s2)

print("Average Investmentment of 1 share = Rs.",round((sumTS/s2),2))



ts_end=1024       #last open price of Tata Steel on 2021-04-30

#calculating investment results

result2=round((ts_end*s2)-sumTS,2)

roiTS=round((result2/sumTS)*100,2)


print("Investment Result:")


if result2<0:

    print("Net Unrealised Loss = Rs.",abs(result2))

else:

    print("Net Unrealised Profit = Rs.",result2)



print("Tata Steel ROI from 2000-1-3 to 2021-04-30 =",roiTS,"%")
```

```python
#TCS ROI

sumTCS=0

s3=0

#calcuating total amount invested and number of shares owned of TCS

for i in range(len(tcs)):

    if tcs.loc[i,'Day']==30:

        sumTCS+=tcs.loc[i,'Open']

        s3+=1

print("Total Invested in TCS = Rs.",round(sumTCS,2))

print("Shares Owned of TCS =",s3)

print("Average Investmentment of 1 share = Rs.",round((sumTCS/s3),2))



tcs_end=3099       #last open price of TCS on 2021-04-30

#calculating investment results

result3=round((tcs_end*s3)-sumTCS,2)

roiTCS=round((result3/sumTCS)*100,2)


print("Investment Result:")


if result3<0:

    print("Net Unrealised Loss = Rs.",abs(result3))

else:

    print("Net Unrealised Profit = Rs.",result3)



print("TCS ROI from 2000-1-3 to 2021-04-30 =",roiTCS,"%")
```

```python

#Investment Results (Graphically)
```

```python
#Plotting ROI on Bar Graph

plt.figure(figsize=(5,7))

stocks=['Tata Motors','Tata Steel','TCS']
ROI=[roiTM,roiTS,roiTCS]
colors=['red','teal','orange']

plt.bar(stocks,ROI,color=colors)
plt.xlabel('Stocks')
plt.ylabel('Percentage (%)')
plt.title('ROI')
```

```python
#Plotting Profit/Loss Amount on Bar Graph

plt.figure(figsize=(5,7))

stocks=['Tata Motors','Tata Steel','TCS']
amt=[result1,result2,result3]
colors=['red','green','orange']

plt.bar(stocks,amt,color=colors)
plt.xlabel('Stocks')
plt.ylabel('Amount (₹)')
plt.title('Profit/Loss')
```

```python
#Portfolio Allocation

plt.figure(figsize=(5,5))

stocks=['Tata Motors','Tata Steel','TCS']
count=[s1,s2,s3]
col=['pink','teal','orange']

plt.pie(count,labels=stocks,autopct="%1.2f%%",colors=col)

plt.legend(title="",loc="upper left")

plt.title("Portfolio Allocation")
```
